```python
import numpy as np

import matplotlib.pyplot as plt

import matplotlib.patches as patches

import keras

from PIL import Image

import math

import copy
```

```python
# Node Class
class Node:
    def __init__(self, position=None, parent=None):
        self.position = position
        self.left_max1 = 0
        self.right_max1 = 0
        self.area_state = 0 # non-rotatable area: 0, fully-rotatable area: 1, partially-rotatable area: 2
        self.parent = parent
        self.cost = 0
        self.cost_rotate = 0
        self.cost_distance_local = 0
        self.cost_distance_global = 0

    def __eq__(self, other):
        return (self.position == other.position)

def GetMapImage(map_path, map_size, map_pixel, global_map_size, global_map_pixel):
  map_raw = Image.open(map_path)
  map_raw = map_raw.resize((map_pixel,map_pixel))
  plt.figure()
  plt.imshow(map_raw, cmap='gray', extent=(0, map_size, 0, map_size))
  plt.xlim(0, map_size)
  plt.ylim(0, map_size)
  plt.show()

  cut_area = (0, 0, global_map_pixel, global_map_pixel)
  map_cut = map_raw.crop(cut_area)
  map_cut = map_cut.resize((global_map_pixel,global_map_pixel))
  plt.figure()
  plt.imshow(map_cut, cmap='gray', extent=(0, global_map_size, 0, global_map_size))
  plt.xlim(0, global_map_size)
  plt.ylim(0, global_map_size)
  plt.show()

  return map_cut


def GetNodes(model, map_cut, edge_dist, global_map_size, local_map_pixel):
  nodes = []

  x = np.delete(np.arange(0, global_map_size, edge_dist), np.where(np.arange(0, global_map_size, edge_dist) == 0))
  y = np.delete(np.arange(0, global_map_size, edge_dist), np.where(np.arange(0, global_map_size, edge_dist) == 0))

  for _y in y:
    for _x in x:
      node = Node((_x,_y))
      nodes.append(node)

  plt.figure()
  plt.imshow(map_cut, cmap='gray', extent=(0, global_map_size, 0, global_map_size))

  for node in nodes:
    plt.scatter(node.position[0], node.position[1], s= 1,c='red', zorder=0)

  plt.xlim(0, global_map_size)
  plt.ylim(0, global_map_size)
  plt.show()

  nodes = GetRotatableArea(nodes, model, map_cut, global_map_size, global_map_pixel, local_map_pixel)

  return nodes

def GetRotatableArea(nodes, model, map_cut, global_map_size, global_map_pixel, local_map_pixel):

  for node in nodes:

    left_upper_x = node.position[0]*global_map_pixel/global_map_size - local_map_pixel/2 #
    left_upper_y = (global_map_size - node.position[1])*global_map_pixel/global_map_size - local_map_pixel/2
    cut_area = (left_upper_x, left_upper_y, left_upper_x + local_map_pixel, left_upper_y + local_map_pixel)
    local_map_raw = map_cut.crop(cut_area)
    local_map_arr = np.array(local_map_raw.resize((local_map_pixel,local_map_pixel)))
    local_map_arr[local_map_arr == 0] = 96.3765
    local_map_arr = local_map_arr.astype('float32')/255
    local_map = local_map_arr.reshape(-1, local_map_pixel, local_map_pixel, 1)

    y_pred = model.predict(local_map, verbose=0)  # 테스트 데이터셋에 대한 예측 결과

    arr_left = np.array(y_pred[0][0])
    arr_right = np.array(y_pred[1][0])
    # print(y_pred)
    print(y_pred[0][0], y_pred[1][0])

    num_left = arr_left.item()  # 첫 번째 요소를 단일 숫자로 변환
    num_right = arr_right.item()

    node.left_max1 = num_left * 190
    node.right_max1 = num_right * 190

    if 183 <= node.left_max1: # non-rotatable area
      node.cost_rotate = 100

    elif 153.5 <= node.left_max1 < 183: # fully-rotatable area
      node.cost_rotate = 0

    elif 0 <= node.left_max1 < 153.5: # partially-rotatable area
      node.cost_rotate = round((180 - (node.left_max1 + node.right_max1))/180, 3)

  return nodes


def GetRotatableMap(nodes, map_cut, global_map_size):

  fig, ax = plt.subplots()
  ax.imshow(map_cut, extent=[0, global_map_size, 0, global_map_size], cmap='gray', zorder=0)

  for node in nodes:

    node_plt_x = node.position[0]
    node_plt_y = node.position[1]

    plt.scatter(node_plt_x, node_plt_y, s= 1, c='red', zorder=1)

    if 183 <= node.left_max1: # non-rotatable area
      node.area_state = 0

      ax.plot(node_plt_x, node_plt_y, 'x', markersize=5, color='red')

    elif 153.5 <= node.left_max1 < 183: # fully-rotatable area
      node.area_state = 1

      fr_pos = (node_plt_x, node_plt_y)
      ax.add_artist(plt.Circle(fr_pos, radius=0.08, color="blue", alpha=0.5, zorder=3))

    elif 0 <= node.left_max1 < 153.5: # partially-rotatable area
      node.area_state = 2

      center = (node_plt_x, node_plt_y)
      radius = 0.08

      start_angle1 = -node.right_max1 + 90
      end_angle1 = node.left_max1 + 90

      wedge1 = patches.Wedge(center, radius, start_angle1, end_angle1, alpha=0.5, fc='blue')
      ax.add_artist(wedge1)

      start_angle2 = start_angle1 + 180
      end_angle2 = end_angle1 + 180

      wedge2 = patches.Wedge(center, radius, start_angle2, end_angle2, alpha=0.5, fc='blue')
      ax.add_artist(wedge2)

  plt.xlim(0, global_map_size)
  plt.ylim(0, global_map_size)
  plt.show()


def VisualizePath(nodes, map_cut, paths, global_map_size):

  fig, ax = plt.subplots()
  ax.imshow(map_cut, extent=[0, global_map_size, 0, global_map_size], cmap='gray', zorder=0)

  for node in nodes:
    node_plt_x = node.position[0]
    node_plt_y = node.position[1]

    plt.scatter(node_plt_x, node_plt_y, s= 1, c='red', zorder=2)

    if 183 <= node.left_max1: # non-rotatable area

      ax.plot(node_plt_x, node_plt_y, 'x', markersize=1, color='red', zorder=1)

      # 20 x 20 map
      # plt.scatter(node_plt_x, node_plt_y, s= 0.5, c='red', zorder=2)

    elif 153.5 <= node.left_max1 < 183: # fully-rotatable area

      fr_pos = (node_plt_x, node_plt_y)
      ax.add_artist(plt.Circle(fr_pos, radius=0.08, color="blue", alpha=0.5, zorder=1))

      # 20 x 20 map
      # plt.scatter(node_plt_x, node_plt_y, s= 0.5, c='blue', zorder=2)

    elif 0 <= node.left_max1 < 153.5: # partially-rotatable area

      center = (node_plt_x, node_plt_y)
      radius = 0.08

      start_angle1 = -node.right_max1 + 90
      end_angle1 = node.left_max1 + 90

      wedge1 = patches.Wedge(center, radius, start_angle1, end_angle1, alpha=0.5, fc='blue', zorder=1)
      ax.add_artist(wedge1)

      start_angle2 = start_angle1 + 180
      end_angle2 = end_angle1 + 180

      wedge2 = patches.Wedge(center, radius, start_angle2, end_angle2, alpha=0.5, fc='blue', zorder=1)
      ax.add_artist(wedge2)

      # 20 x 20 map
      # plt.scatter(node_plt_x, node_plt_y, s= 0.5, c='blue', zorder=2)

  ax.plot([path.position[0] for path in paths], [path.position[1] for path in paths], color="yellow", linewidth=1.0, zorder=3)

  plt.xlim(0, global_map_size)
  plt.ylim(0, global_map_size)
  plt.show()


def Astar(nodes, start, target, edge_dist, global_map_size, NofPoint, output, weight_r, weight_dl):

  openList = []
  closedList = []
  cnt = 0

  startIdx = round((start[0]/edge_dist - 1) + (start[1]/edge_dist - 1)*NofPoint)
  targetIdx = round((target[0]/edge_dist - 1) + (target[1]/edge_dist - 1)*NofPoint)

  startNode = copy.deepcopy(nodes[startIdx])
  targetNode = copy.deepcopy(nodes[targetIdx])

  openList.append(startNode)

  # targetNode를 찾을 때까지 실행
  # openList가 비어있으면 반복을 종료
  while openList:

    # 현재 노드 지정
    currentNode = openList[0]
    currentIdx = 0

    # openList 중에서 f가 가장 작은 node 찾기
    for index, item in enumerate(openList):
      if item.cost < currentNode.cost:
        currentNode = item
        currentIdx = index

    # openList에서 제거하고 closedList에 추가
    openList.pop(currentIdx)
    closedList.append(currentNode)

    if output == 1:
      for closednode in closedList:
        if closednode.parent is not None:
          print("closed node : (", round(closednode.position[0], 3),",",round(closednode.position[1], 3),")"," parent : (", round(closednode.parent.position[0], 3),",",round(closednode.parent.position[1], 3),")")
        else:
          print("closed node : (", round(closednode.position[0], 3),",",round(closednode.position[1], 3),")")

      cnt += 1
      print("==================================")
      print(cnt)
      if currentNode.parent is not None:
        print("current node : (", round(currentNode.position[0], 3),",",round(currentNode.position[1],3),")"," parent : (", round(currentNode.parent.position[0], 3),",",round(currentNode.parent.position[1],3),")")
      else:
        print("current node : (", round(currentNode.position[0], 3),",",round(currentNode.position[1],3),")")
      if currentNode.parent is not None:
        print("current node parent : (", round(currentNode.parent.position[0], 3),",",round(currentNode.parent.position[1],3),")")

    # 현재 노드가 목적지면 current.position 추가하고
    # current의 부모로 이동
    if currentNode == targetNode:
      path = []
      current = currentNode
      print("=================================")
      print("find path!")
      while current is not None:
        print("path : (", round(current.position[0], 3),",",round(current.position[1],3),")")
        path.append(current)
        current = current.parent
      return path[::-1]  # reverse

    children = []
    candidates = [(0, -edge_dist), (0, edge_dist), (-edge_dist, 0), (edge_dist, 0), (-edge_dist, -edge_dist), (-edge_dist, edge_dist), (edge_dist, -edge_dist), (edge_dist, edge_dist)]

    # 인접한 xy좌표 전부
    for nearPosition in candidates:
      nearPosition_x = currentNode.position[0] + nearPosition[0]
      nearPosition_y = currentNode.position[1] + nearPosition[1]

      # 미로 maze index 범위 안에 있어야함
      within_range_criteria = [
        nearPosition_x >= global_map_size,
        nearPosition_x <= 0,
        nearPosition_y >= global_map_size,
        nearPosition_y <= 0,
      ]

      # 범위 밖이면 자식 x
      if any(within_range_criteria):  # 하나라도 true면 범위 밖임
        continue

      # near node가 non-rotatable area면 자식 x
      nearIdx = round((nearPosition_x/edge_dist - 1) + (nearPosition_y/edge_dist - 1)*NofPoint)
      nearNode = copy.deepcopy(nodes[nearIdx])

      # nearNode = Node((nearPosition_x, nearPosition_y), currentNode)

      if nearNode.area_state == 0:
        continue

      if nearPosition == candidates[4] or nearPosition == candidates[5] or nearPosition == candidates[6] or nearPosition == candidates[7]:
        nearNode.cost_distance_local = currentNode.cost_distance_local + math.sqrt(edge_dist)
      else:
        nearNode.cost_distance_local = currentNode.cost_distance_local + math.sqrt(edge_dist**2 + edge_dist**2)

      # near node가 closedList에 있으면 자식 x
      # in은 값 비교(주솟값이 아니라)
      if nearNode in closedList:
        # print("nearNode in closedList : (",round(nearNode.position[0],3),",", round(nearNode.position[1],3),")")
        continue

      children.append(nearNode)

    # 자식들 모두 loop
    for child in children:

      child.parent = currentNode

      # f, g, h값 업데이트
      if currentNode.area_state == 1: # fully-rotatable area
          if child.area_state == 1: # fully-rotatable area
            child.cost_rotation = 0
          elif child.area_state == 2:
            child.cost_rotate = round((180 - (child.left_max1 + child.right_max1))/180, 3)
      elif currentNode.area_state == 2: # partially-rotatable area
        if child.area_state == 1: # fully-rotatable area
          child.cost_rotate = round((180 - (currentNode.left_max1 + currentNode.right_max1))/180, 3)
        elif child.area_state == 2: # partially-rotatable area
          if find_intersection(child.left_max1, child.right_max1, currentNode.left_max1, currentNode.right_max1) is not None: # intersection O
            left_max1, right_max1 = find_intersection(child.left_max1, child.right_max1, currentNode.left_max1, currentNode.right_max1)
            child.cost_rotate = round((180 - (child.left_max1 + child.right_max1))/180, 3)
          else: # intersection X
            child.cost_rotate = 100

      # child.cost_distance_global = math.sqrt((child.position[0] - targetNode.position[0])**2 + (child.position[1] - targetNode.position[1])**2)/2*(global_map_size*math.sqrt(2))

      child.cost_distance_global = math.sqrt((child.position[0] - targetNode.position[0])**2 + (child.position[1] - targetNode.position[1])**2)

      weight_dg = 1 - weight_r - weight_dl
      # weight_dl = 1
      # weight_dg = 1

      child.cost = weight_r * child.cost_rotate + weight_dl * child.cost_distance_local + weight_dg * child.cost_distance_global

      # weight_dg = 1 - weight_r
      # child.cost = weight_r * child.cost_rotate + weight_dg * child.cost_distance_global

      if output == 1:
        print("child node : (",round(child.position[0],3),",", round(child.position[1],3),")", ", g : ", round(child.cost_rotate, 3), ", h : ", round(child.cost_distance, 3), ", c : ", round(child.cost, 3))

      list = [(index, openNode) for index, openNode in enumerate(openList) if child == openNode]

      if len(list) == 0:
        openList.append(child)
      else:
        tuple = list[0]
        index = tuple[0]
        openNode = tuple[1]
        if child.cost_rotate >= openNode.cost_rotate:
          continue
        else:
          openList.pop(index)
          openList.append(child)

    if output == 1:
      for opennode in openList:
        if opennode.parent is not None:
          print("open node : (", round(opennode.position[0], 3),",",round(opennode.position[1], 3),")", ", g : ", round(opennode.cost_rotate, 3), ", h : ", round(opennode.cost_distance, 3), ", c : ", round(opennode.cost, 3)," parent : (", round(opennode.parent.position[0], 3),",",round(opennode.parent.position[1], 3),")")
        else:
          print("open node : (", round(opennode.position[0], 3),",",round(opennode.position[1], 3),")", ", g : ", round(opennode.cost_rotate, 3), ", h : ", round(opennode.cost_distance, 3), ", c : ", round(opennode.cost, 3))

def find_intersection(_left_yaw_max1, _right_yaw_max1, _left_yaw_max2, _right_yaw_max2):
    left_yaw_max1 = _left_yaw_max1 + 90
    right_yaw_max1 = -_right_yaw_max1 + 90
    left_yaw_max2 = _left_yaw_max2 + 90
    right_yaw_max2 = -_right_yaw_max2 + 90

    start = min(left_yaw_max1, left_yaw_max2)
    end = max(right_yaw_max1, right_yaw_max2)

    start = start - 90
    end = -end + 90

    if start <= end:
        return start, end
    else:
        return None
```

```python
# load map

m2p = 50 # pixel:m 에 맞게 조정

map_size = 20
map_pixel = map_size*m2p #1000
global_map_size = 5 #20
global_map_pixel = int(round(global_map_size*m2p, 0)) #250
local_map_size = 1.4
local_map_pixel = int(round(local_map_size*m2p, 0)) #70
edge_dist = 0.2
edge_pixel = int(round(edge_dist*m2p, 0)) #10
NofPoint = int(round(global_map_size/edge_dist, 0) - 1) # 한 행에 속한 point의 갯수
print(NofPoint)

map_cut = GetMapImage("../global_map/terrain_3.png", map_size, map_pixel, global_map_size, global_map_pixel)

# load CNN model
import h5py
# file_path = '../models/20230613(4).h5'
file_path = '../models/20230627(1)_cpu.h5'
model = keras.models.load_model(file_path)

nodes = GetNodes(model, map_cut, edge_dist, global_map_size, local_map_pixel)
GetRotatableMap(nodes, map_cut, global_map_size)
print("done")
```
